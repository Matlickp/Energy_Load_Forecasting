```python
import pandas as pd
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.energyDB
coll = db.forecastSDGE
coll2 = db.forecastSDGEdates
```

```python
forecast_df = pd.read_csv('FCWeatherdata_SDGE.csv') 
forecast_df.head()
```

```python
date_col = []
hour_col = []

for index, row in forecast_df.iterrows():
    date_col.append(row['FCDATE'].split()[0])
    hour_col.append(row['FCDATE'].split()[1])
    
forecast_df['DATE'] = date_col
forecast_df['HOUR'] = hour_col
forecast_df
```

```python
forecast_df
```

```python
forecast_df['CODE'].unique()
```

```python
forecast_dict = forecast_df.to_dict(orient='records')
forecast_dict
```

```python
coll.remove()

for record in forecast_dict:
    coll.insert_one(record)
```

```python
datelist = forecast_df['DATE'].unique().tolist()
datelist
```

```python
for date in datelist:
    coll2.insert_one({'DATE': date})
```
